# Simulate closed-loop
After implementing your control functionality, you can simulate the closed-loop with code that looks something like this...

```python
import sys
import os

# Add the directory containing uuv_mission to the Python path
sys.path.append(os.path.abspath('/Users/luke/Documents/Uni/Eng/Year 3/B1/Scientific Coding/b1-coding-practical-mt24'))

# Import relevant modules
import numpy as np
import pandas as pd

# Import classes from uuv_mission
from uuv_mission.dynamic import Submarine, ClosedLoop, Mission
from uuv_mission.controller import PDController

# Instantiate the Submarine
sub = Submarine()

# Instantiate your controller
controller = PDController()

# Create a ClosedLoop instance
closed_loop = ClosedLoop(sub, controller)

# Construct the path to the mission.csv file
current_dir = os.getcwd()
csv_path = os.path.abspath(os.path.join(current_dir, '..', 'data', 'mission.csv'))

# Load the mission from the CSV file
mission = Mission.from_csv(csv_path)

# Simulate with random disturbances
trajectory = closed_loop.simulate_with_random_disturbances(mission)

# Plot the completed mission
trajectory.plot_completed_mission(mission)
```
